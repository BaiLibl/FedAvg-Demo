```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd

def read_file(f):
    y = []
    f = open(f, 'r')
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        # print(line)
        y.append(float(line[0:5]))
    return y

f1 = 'cnn_cifar_100_loss.txt'
f2 = 'cnn_cifar_100_acc.txt'

loss = read_file(f1)
acc = read_file(f2)
x = [(i+1)*2 for i in range(len(loss))]

plt.style.use('seaborn')
fig = plt.figure()

ax = fig.add_subplot(111)
lns1 = ax.plot(x, acc, ':s', color='#03C4A1', label=r'Acc', linewidth=3.0,markersize=10.0)
plt.legend(frameon=False, prop={'size':25},loc='lower right')
ax2 = ax.twinx()
lns2 = ax2.plot(x, loss, '-.o', color='#C62A88', label=r'Loss',linewidth=3.0,markersize=10.0)

plt.legend(bbox_to_anchor=(1.02, 0.35),frameon=False,prop={'size':25})
plt.xticks(size=20)

# ax.grid()
ax.set_xlabel('Epoch', fontsize=20)
ax.set_ylabel('Train Accuracy',fontsize=20)
ax.yaxis.set_tick_params(labelsize=15)
ax.xaxis.set_tick_params(labelsize=15)
ax2.set_ylabel('Train Loss',fontsize=20)
ax2.yaxis.set_tick_params(labelsize=15)

plt.savefig('cnn_cifar_100_train.png')

plt.show()
```
